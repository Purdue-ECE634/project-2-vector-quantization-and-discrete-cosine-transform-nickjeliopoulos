```python
import torch
import treescope
treescope.basic_interactive_setup()
```

```python
B = 4
C, H, W = 1, 16, 16
L = H // B * W // B
image = torch.randn(size=[H, W, C])
image
```

```python
def blockify_image(image: torch.Tensor) -> torch.Tensor:
	H, W, C = image.shape
	blocks = image.unfold(0, B, B).unfold(1, B, B)
	blocks = blocks.reshape(-1, B * B, C)
	return blocks

def unblockify_image(blocks: torch.Tensor, image_shape) -> torch.Tensor:
	H, W, C = image_shape
	blocks = blocks.view(H // B, W // B, B, B, C)
	blocks = blocks.permute(0, 2, 1, 3, 4).contiguous()
	return blocks.view(H, W, C)
```

```python
blocks = blockify_image(image)
# blocks
```

```python
reconstruct = unblockify_image(blocks, image.shape)
reconstruct
```

```python
torch.allclose(image, reconstruct)
```
